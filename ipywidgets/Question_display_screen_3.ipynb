```python
def construct_qbox(qdetails):
    global review_question_stem_dict,central_concept,conceptcluster
#     lenq = len([k for k in review_question_stem_dict[q_key].keys()])
    # select one at random
#     keyn = random.choice(list(range(lenq)))
    # get length of questions for this cue
    question_string = ''
    # there is a better more pythonic way to do the loop below, surely
    for s in qdetails['prompt']:
        if s=='central_concept':
            question_string = question_string + central_concept
        else:
            question_string = question_string + s
    question_prompt= widgets.HTML(question_string)
    if type(qdetails['answer_options'])==list:
        answer_options = qdetails['answer_options']
    if qdetails['answer_options']=='neighbors':
        answer_options= [a for a in conceptcluster if a is not central_concept]
        answer_options.sort()
        answer_options = answer_options+ ['None of the above']
    question_style = qdetails['question_style']
    if question_style=='RadioButton':
        answer_prompt = widgets.RadioButtons(options=answer_options)
    elif question_style=='DropDown':
        answer_prompt = widgets.DropDown(options=answer_options)
    elif question_style=='SelectMultiple':
        answer_prompt = widgets.SelectMultiple(options=answer_options)
    elif question_style=='Combobox':
        answer_prompt = widgets.Combobox(options=answer_options,ensure_options=False)
    elif question_style=='FloatText':
        answer_prompt = widgets.FloatText(value=answer_options)
    elif question_style=='TextArea':
        answer_prompt = widgets.Textarea(value='',disabled=False)

#     elif question_style='CheckboxMultiple':
#         vboxlist = [ans_opt for ans_opt in answer_options]
            
    return (question_prompt,answer_prompt)

# def store_answer(question_string,answer_saved,central_concept):
#     global review_question_stem_dict, qa_history, conceptcuedict
#     # history list of strings adds this
#     qa_history = qa_history + [question_string+': ',answer_saved]
# #     conceptcuedict[central_concept] = 
#     # store answer in database
#     if review_question_stem_dict[]
    
# def get_next_q(answer_cue)
```

```python
def update_question_thread():
    global conceptcuedict, central_concept,qcue,qdetails, answer_saved, question_cue, conceptcluster
    if type(qdetails['predicate_storage'])==dict:
        predicate = qdetails['predicate_storage'].get(answer_saved,'')
    else:
        predicate = qdetails['predicate_storage']
    # here would be storing answer, running inferences, finding contradictions/updating conceptcuedict for each concept, then changing cognitive laod calculations for the hints too
    # assuming all that is done, then
    # over here we would trim subject concept and object concept for roles already eliminated
    conceptcuedict[central_concept] = conceptcuedict[central_concept] + [i for i in question_cue.get(predicate,[]) if i not in conceptcuedict[central_concept]]
    curr_concept = central_concept
    if len(conceptcuedict[curr_concept])<=1:
        central_concept = find_next_central_concept(curr_concept)
    if answer_saved!='No answer saved':
        conceptcuedict[curr_concept].remove(qcue)
    else:
        conceptcuedict[curr_concept].remove(qcue)
        conceptcuedict[curr_concept] = conceptcuedict[curr_concept]+ [qcue]
    if central_concept=='':
        qcue = 'stop'
    # if question was answered, first, add question items to the current central_concept based on predicate
    # check the length of the conceptcuedict[central_concept]
    # if it only contains current qcue, then if question was answered, remove and find the next central_concept with a cue list >=1
    # if none exists, return stop qcue value
    # if it only contains current qcue, then if question was skipped, find next central_concept with a cue list
    # if none exists, return stop qcue value
def find_next_central_concept(central_concept):
    global conceptcuedict
    concepts_left= [c for c in conceptcuedict.keys() if (c!=central_concept) & (len(conceptcuedict[c])>0)]
    if len(concepts_left)>0:
        return concepts_left[0]
```
